```python
# Smart Device Management System

class SmartDevice:
    def __init__(self, name, device_id):
        self.name = name
        self.__device_id = device_id
        self.__power = False

    # Getters
    def get_device_id(self):
        return self.__device_id

    def get_power(self):
        return self.__power

    # Setters
    def set_device_id(self, device_id):
        if device_id:
            self.__device_id = device_id
        else:
            print("Device ID cannot be empty.")

    def turn_on(self):
        self.__power = True
        print(f"{self.name} is ON.")

    def turn_off(self):
        self.__power = False
        print(f"{self.name} is OFF.")

    def display_info(self):
        print(f"\nDevice: {self.name}")
        print(f"ID: {self.__device_id}")
        print(f"Power: {'ON' if self.__power else 'OFF'}")


class SmartLight(SmartDevice):
    def __init__(self, name, device_id, brightness):
        super().__init__(name, device_id)
        self.brightness = brightness

    def increase_brightness(self):
        if self.brightness < 200:
            self.brightness = self.brightness + 20
        print("Brightness:", self.brightness)

    def decrease_brightness(self):
        if self.brightness > 50:
            self.brightness = self.brightness - 10
        print("Brightness:", self.brightness)

    def display_info(self):
        super().display_info()
        print("Brightness:", self.brightness)


class SecurityCamera(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.recording = False

    def start_recording(self):
        self.recording = True
        print("Recording began.")

    def stop_recording(self):
        self.recording = False
        print("Recording ended.")

    def display_info(self):
        super().display_info()
        print("Recording:", self.recording)


class TemperatureSensor(SmartDevice):
    def __init__(self, name, device_id, temperature):
        super().__init__(name, device_id)
        self.temperature = temperature

    def read_temperature(self):
        print("Temperature:", self.temperature, "°C")

    def display_info(self):
        super().display_info()
        print("Temperature:", self.temperature, "°C")


# Objects
light = SmartLight("Bathroom ligth", "R021", 80)
camera = SecurityCamera("Kitchen camera", "C342")
sensor = TemperatureSensor("Door Sensor", "D7302", 31)

while True:
    print("\n=== Smart Device Menu ===")
    print("1. Display Information")
    print("2. Turn Devices ON")
    print("3. Turn Devices OFF")
    print("4. Read Temperature")
    print("5. Adjust Brightness")
    print("6. Recording")
    print("7. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        light.display_info()
        camera.display_info()
        sensor.display_info()

    elif choice == "2":
        light.turn_on()
        camera.turn_on()
        sensor.turn_on()

    elif choice == "3":
        light.turn_off()
        camera.turn_off()
        sensor.turn_off()

    elif choice == "4":
        sensor.read_temperature()

    elif choice == "5":
        option = input("1. Increase\n2. Decrease\nChoose: ")
        if option == "1":
            light.increase_brightness()
        elif option == "2":
            light.decrease_brightness()

    elif choice == "6":
        option = input("1. Start\n2. Stop\nChoose: ")
        if option == "1":
            camera.start_recording()
        elif option == "2":
            camera.stop_recording()

    elif choice == "7":
        print("Bye bye!")
        break

    else:
        print("Error.")
```
